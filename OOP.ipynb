Question 1: 

Create a base class named CollegeMember with:

Attributes: name, id_number

Method: display_info() -> prints name and ID number.

Then create a derived class Student with:

Additional attribute: program (e.g., "BISI", "CS", "Nursing")

Override the display_info() method to include program.


```python
class CollegeMember():
    def __init__(self,name,id_number):
        self.name = name
        self.id_number = id_number

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Id number: {self.id_number}")

class Student(CollegeMember):
    def __init__(self,name,id_number,program):
        self.name = name
        self.id_number = id_number
        self.program = program

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Id number: {self.id_number}")
        print(f"Program: {self.program}")
```


Task:

```python
student2 = Student("Bob",102,"CS")
student2.display_info()
```

Question 2: 

Extend the Student class to a new class InternationalStudent which has:

Additional attribute: country_of_origin

Method: welcome_message() → prints:
"Welcome <name> from <country>, enrolled in <program> program."



```python
class InternationalStudent(Student):
    def __init__(self,name,id_number,program,country_of_origin):
        self.name = name
        self.id_number = id_number
        self.program = program
        self.country_of_origin = country_of_origin

    def welcome_message(self):
        print(f"Welcome {self.name} from {self.country_of_origin}, enrolled in {self.program} program.")
```

Question 3: 

Create a base class CourseFee with method calculate_fee() that returns base fee = 2000.

Create two subclasses:

UndergraduateFee (adds $500 lab fee)

GraduateFee (adds $1200 research fee)

Use polymorphism by writing a loop to call calculate_fee() for:

```python
class CourseFee():

    def calculate_fee(self):
        return 2000

class UndergraduateFee(CourseFee):

    def calculate_fee(self):
        base_fee = super().calculate_fee()
        lab_fee = 500

        return base_fee + lab_fee

class GraduateFee(CourseFee):

    def calculate_fee(self):
        base_fee = super().calculate_fee()
        res_fee = 1200

        return base_fee + res_fee
```

Task:
Print calculated fees for both.

```python
und1 = UndergraduateFee()
grad1 = GraduateFee()

for fee in [und1,grad1]:
    print(f"Total fee is ${fee.calculate_fee()}.")
```

Question 4: 

Create three classes: Teacher, Student, and Administrator.
Each class must have a method: role_description()

Teacher → “I teach courses and evaluate students.”

Student → “I attend classes and complete assignments.”

Administrator → “I manage campus operations.”

```python
class Teacher:

    def role_description(self):
        return 'I teach courses and evaluate students.'

class Student:

    def role_description(self):
        return 'I attend classes and complete assignments.'

class Administrator:
    
    def role_description(self):
        return 'I manage campus operations.'
```

Task:
Create objects of each class and call role_description() using a loop to demonstrate polymorphism.

```python
teacher1 = Teacher()
student3 = Student()
administrator1 = Administrator()
```

```python
for college_member1 in [teacher1, student3, administrator1]:
    print(college_member1.role_description())
```

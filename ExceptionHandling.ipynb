Handle the exception thrown by the code below by using <code>try</code> and <code>except</code> blocks.

Write a Python program to handle a ZeroDivisionError exception when dividing a number by zero.

```python
def math_division(a,b):

    try:
        result = a / b
        print(result)
    except ZeroDivisionError:
        print('Error : Dividing a number by zero.')
```

### Problem 4

Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.

exception TypeError:

Raised when an operation or function is applied to an object of inappropriate type. The associated value is a string giving details about the type mismatch.

```python
def two_numbers():
    try:
        s1 = input("Please input a number : ")
        s2 = input("Please input a second number : ")

        if (s1.isdigit() and s2.isdigit()):
            s1_1 = float(s1)
            s2_1 = float(s2)
        else:
            raise TypeError('Error : The inputs are not numerical.')
          
        result = s1_1 + s2_1
        print(f"The sum of the two inputs is {result}")
    except TypeError as e:
        print(f"{e}")
        
```

```python
two_numbers()
```

```python
two_numbers()
```

### Problem 5
Write a function that asks for an integer and prints the square of it. Use a <code>while</code> loop with a <code>try</code>, <code>except</code>, <code>else</code> block to account for incorrect inputs.

```python
def func_squared():
    while True:
        try:
            value = int(input("Please input an integer : "))
            result = value ** 2
        except ValueError:
            print("Error : An integer was not input.")
        else: 
            print("An integer was input.")
            print(f"{value} squared is {result}")
            break
```

```python
func_squared()        
```

### Problem 6

Write a Python program that executes an operation on a list and handles an IndexError exception if the index is out of range.

exception IndexError:

Raised when a sequence subscript is out of range. (Slice indices are silently truncated to fall in the allowed range; if an index is not an integer, TypeError is raised.)

```python
def func_times_two(a):
    try:
        for i in range(0,len(a)+1):
            a[i] *= 2
            print(a[i])
        print(f"Each element of the list multiplied by 2 is : {a}")
    except IndexError:
        print('Error: The index is out of range.')
    
```

```python
func_times_two([1,3,5,7,9])
```

### Problem 7

Write a Python program that executes division and handles an ArithmeticError exception if there is an arithmetic error.

exception ArithmeticError:

The base class for those built-in exceptions that are raised for various arithmetic errors: OverflowError, ZeroDivisionError, FloatingPointError.

```python
def func_division(a,b):
    try:
        result = a / b
        print(f"The result of the division is {result}")
    except ArithmeticError:
        print('Error : An arithmetic error occured.')
```

```python
func_division(50 ** 10000,1)
```

```python
func_division(80,0)
```

```python
func_division(8,2)
```

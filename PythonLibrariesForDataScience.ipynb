### **Tasks 1:**

#### 1.	Import NumPy as np.
#### 2.	Create a 2D array (5x5) with random integers between 10 and 50.
#### 3.	Find:
#### -	The mean, median, and standard deviation.
#### -	The maximum and minimum elements and their indices.


```python
import numpy as np
val = np.random.randint(10,51,size=(5,5))
print(val)
```

```python
# calculate mean
val_mean = np.mean(val)
print(val_mean)
```

```python
# calculate median
val_median = np.median(val)
print(val_median)
```

```python
# calculate standard deviation
val_std = np.std(val)
print(val_std)
```

```python
# find the max element
val_max = np.max(val)
print(val_max)
```

```python
val_flattened = val.ravel()
val_max_index_flattened = val_flattened.argmax()
val_max_index_2d = np.unravel_index(val_max_index_flattened,val.shape)

print(f"Max value is {val_max} at row = {val_max_index_2d[0]} and col = {val_max_index_2d[1]}.")
```

```python
# find the min element
val_min = np.min(val)
print(val_min)
```

```python
val_flattened = val.ravel()
val_min_index_flattened = val_flattened.argmin()
val_min_index_2d = np.unravel_index(val_min_index_flattened,val.shape)

print(f"Min value is {val_min} at row = {val_min_index_2d[0]} and col = {val_min_index_2d[1]}.")
```

### **Tasks 2:**

#### 1.	Import scipy.stats and scipy.integrate.
#### 2.	Create a dataset of 100 random numbers from a normal distribution with mean=50 and std=10.
#### 3.	Compute:
#### -	Mean, median, mode, variance, skewness, and kurtosis.


```python
import scipy.stats as scs
import scipy.integrate

val2 = scs.norm.rvs(loc=50,scale=10,size=100)
print(val2)
```

```python
# compute the mean, mode, median, variance, skewness, and kurtosis
val2_mean = np.mean(val2)
val2_mode = scs.mode(val2)
val2_median = np.median(val2)
val2_var = np.var(val2, ddof=1)
val2_skew = scs.skew(val2)
val2_kurt = scs.kurtosis(val2)
```

```python
print(val2_mean)
```

```python
print(val2_mode[0])
```

```python
print(val2_median)
```

```python
print(val2_var)
```

```python
print(val2_skew)
```

```python
print(val2_kurt)
```

### **Tasks 4:**

#### 1.	Plot:
#### -	Line graph for monthly sales.
#### - 	Bar chart for category comparison.
#### -	Pie chart for department-wise employee count.


```python
import matplotlib.pyplot as plt

data2 = {
'month' : ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
'sales' : [10000, 10500, 11000, 12000, 9800, 13000, 13500, 15000, 17000, 18000, 14500, 20000],
'category' : ['Clothing', 'Clothing', 'Clothing', 'Clothing', 'Groceries', 'Groceries', 
              'Groceries', 'Groceries', 'Groceries', 'Groceries', 'Electronics', 'Electronics']
}

df2 = pd.DataFrame(data2)

# line plot of monthly sales

plt.plot(df2['month'],df2['sales'])
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
category_count = df2['category'].value_counts()

# bar chart for category comparison

plt.bar(category_count.index, category_count.values)
plt.title('Count of Each Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
dep_count = df['Department'].value_counts() 

# pie chart for department-wise employee count

plt.pie(dep_count.values, autopct = '%.1f%%', radius = 1.2, labels = dep_count.index)
plt.title('Department-Wise Employee Count')
plt.show()
```
